```python
# Standard python libraries
import os
import requests

# Essential DS libraries
import numpy as np
import pandas as pd
```

```python
sales_agg_features_target = pd.read_csv('./data/sales_agg_features_target.csv',sep=';').drop(['target'],axis=1).rename(columns={'ID':'id_shop_item'})
sales_agg_features_target_sub = sales_agg_features_target[sales_agg_features_target['date_block_num']==33]

sales_agg_features_target_sub.head(10)
```

```python
sales_agg_features_target_sub['item_category_id'] = sales_agg_features_target_sub['item_category_id'].astype(str)
sales_agg_features_target_sub['date_block_num'] = sales_agg_features_target_sub['date_block_num'].astype(str)
```

```python
sales_agg_features_target_sub.dtypes
```

```python
submission = pd.read_csv('./data/test.csv',sep=',',dtype=object)
submission
```

```python
submission['id_shop_item'] = submission['shop_id'] + '_' + submission['item_id']
submission = submission.merge(sales_agg_features_target_sub,on='id_shop_item',how='left')
submission.head(10)
```

```python
submission.info()
```

```python
sales_agg_features_target_sub_agg = sales_agg_features_target.groupby(by='id_shop_item'). \
    agg({'item_category_id':'max', 'date_block_num':'max', 'item_price_sum':'mean', 'item_price_mean':'mean', \
        'item_cnt_day_sum':'mean', 'item_cnt_day_mean':'mean'})
sales_agg_features_target_sub_agg
```

```python
sales_agg_features_target_sub_agg = sales_agg_features_target_sub_agg.reset_index()
sales_agg_features_target_sub_agg.head(10)
```

```python
submission_na = submission[submission['item_category_id'].isna()]
submission_na = submission_na[['ID', 'shop_id',	'item_id','id_shop_item']].merge(sales_agg_features_target_sub_agg,on='id_shop_item',how='left')
submission_na.info()
```

```python
sales_agg_features_target_sub['item_id'] = sales_agg_features_target_sub['id_shop_item'].apply(lambda x: x.split('_')[1])
sales_agg_features_target_sub.head(5)
```

```python
sales_agg_features_target_sub_agg = sales_agg_features_target_sub.drop(['id_shop_item'],axis=1).groupby(by='item_id'). \
    agg({'item_category_id':'max', 'date_block_num':'max', 'item_price_sum':'mean', 'item_price_mean':'mean', \
        'item_cnt_day_sum':'mean', 'item_cnt_day_mean':'mean'})
sales_agg_features_target_sub_agg
```

```python
sales_agg_features_target_sub_agg = sales_agg_features_target_sub_agg.reset_index()
sales_agg_features_target_sub_agg.head(5)
```

```python
submission_na_na = submission_na[submission_na['item_category_id'].isna()]
submission_na = submission_na[~submission_na['item_category_id'].isna()]
submission_na_na = submission_na_na[['ID', 'shop_id',	'item_id']].merge(sales_agg_features_target_sub_agg,on='item_id',how='left')
submission_na_na.info()
```

```python
submission_na_na['item_category_id'] = submission_na_na['item_category_id'].fillna(str(submission_na_na['item_category_id'].value_counts().argmax()))
submission_na_na['date_block_num'] = submission_na_na['date_block_num'].fillna(str(submission_na_na['date_block_num'].value_counts().argmax()))
submission_na_na['item_price_sum'] = submission_na_na['item_price_sum'].fillna(submission_na_na['item_price_sum'].mean())
submission_na_na['item_price_mean'] = submission_na_na['item_price_mean'].fillna(submission_na_na['item_price_mean'].mean())
submission_na_na['item_cnt_day_sum'] = submission_na_na['item_cnt_day_sum'].fillna(submission_na_na['item_cnt_day_sum'].mean())
submission_na_na['item_cnt_day_mean'] = submission_na_na['item_cnt_day_mean'].fillna(submission_na_na['item_cnt_day_mean'].mean())

submission_na_na.info()
```

```python
submission_0 = submission[~submission['item_category_id'].isna()]
submission_0.head(10)
```

```python
submission_0 = submission_0.drop(['shop_id','item_id','id_shop_item'],axis=1)
submission_0['item_category_id'] = submission_0['item_category_id'].astype(str)
submission_0['date_block_num'] = submission_0['date_block_num'].astype(str)

submission_na = submission_na.drop(['shop_id','item_id','id_shop_item'],axis=1)
submission_na['item_category_id'] = submission_na['item_category_id'].astype(str)
submission_na['date_block_num'] = submission_na['date_block_num'].astype(str)

```

```python
submission_na_na = submission_na_na.drop(['shop_id','item_id'],axis=1)
submission_na_na['item_category_id'] = submission_na_na['item_category_id'].astype(str)
submission_na_na['date_block_num'] = submission_na_na['date_block_num'].astype(str)
```

```python

submission_preprocessed = pd.concat((submission_0,submission_na, submission_na_na),axis=0)
submission_preprocessed['ID'] = submission_preprocessed['ID'].astype(int)
submission_preprocessed = submission_preprocessed.sort_values(by='ID',ascending=True)
submission_preprocessed.head(10)
```

```python
submission_preprocessed.dtypes
```

```python
submission_preprocessed.to_csv('./data/submission_preprocessed.csv',sep=';',index=False)
```
